```python
import pandas as pd
import itertools
import matplotlib.pyplot as plt
df = pd.read_csv("Iris_1.csv")
```

```python
df = df.drop('Id', axis=1)
```

```python
df = df.dropna().drop_duplicates()
```

```python
df = df[df['SepalLengthCm'] > 0]
```

```python
df['Species'].value_counts()
df = df[df['Species'] != "Iris-Imsa"]
df
```

```python
df.describe()
```

```python
df.corr()
# most of these have very high positive pearson's r relations
```

```python
cols = list(df.corr().columns)[1:]
combos = list(itertools.combinations(cols, 2))
```

```python
plt.figure(figsize=(27, len(combos) + 1))
plt.suptitle("Iris Data Overview")
basis = 100 + 10 * len(combos)
for (x, y) in combos:
  basis += 1
  plt.subplot(basis)
  plt.scatter(list(df[x]), list(df[y]))
  plt.title(x + ' vs. ' + y)
  plt.xlabel(x)
  plt.ylabel(y)
  
```
