```python
# Dev Singh and Grayson Pacourek
import sys
import json
import matplotlib.pyplot as plt

LABELS = ["Low Reviews", "Medium Reviews", "High Reviews"]

def load_dictionary_from_json(filename):
    f = open(filename)
    data = json.load(f)
    
    f.close
    return data

def plot_women_words(word_data, word, max_frequency):
    plot1 = plt.figure(1)
    x_vals = [0, 1, 2]
    
    y_vals = word_data[word]["W"][2]
    plt.bar(x_vals, y_vals, color="tab:pink")
    plt.title("Female frequncy of '" + word + "' in high reviews")
    plt.ylim((0, max_frequency * 1.5))
    plt.xticks(x_vals, LABELS) 

def plot_man_words(word_data, word, max_frequency):
    plot2 = plt.figure(2)
    x_vals = [0, 1, 2]


    y_vals = word_data[word]["M"][2]
    plt.bar(x_vals, y_vals, color="tab:blue")
    plt.title("Male frequncy of '" + word + "' in high reviews")

    plt.ylim((0, max_frequency * 1.5)) 
    plt.xticks(x_vals, LABELS) 
    plt.show()
```

```python
word_data = load_dictionary_from_json('word_data.json')
word = "daydreaming"
gender_data = word_data[word]
max_frequency = max(max(gender_data['W']), max(gender_data['M']))
print(max(gender_data['W']), max(gender_data['M']))
plot_women_words(word_data, word, max_frequency)
plot_man_words(word_data, word, max_frequency)
```

```python
def get_change(current, previous):
    if current == previous:
        return 0
    try:
        return (abs(current - previous) / previous) * 100.0
    except ZeroDivisionError:
        return 0
```

```python
import random
i = 0
l = list(word_data.keys())
random.shuffle(l)
for word in l:
    if i == 20:
        break
    gender_data = word_data[word]
    w, m = max(gender_data['W']), max(gender_data['M'])
    if get_change(w, m) > 50:
        plot_women_words(word_data, word, max_frequency)
        plot_man_words(word_data, word, max_frequency)
    i += 1
```
